* `` * `` --> Eleman bazlı işlem yapar
* `` @ `` --> Eğer değişken bir vektörse (``np.array([])``) iç çarpım eğer değişken matrisse (``np.array([[]])``) matris çarpımı yapar

````python
A = np.array([[1, 2, 3],
              [1, 2, 3],
              [1, 2, 3]]) 

B = np.array([[1, 2, 3],
              [1, 2, 3],
              [1, 2, 3]])

x = A * B  # eleman - eleman çarpma işlemi
y = A @ B  # matris çarpımı - satır x sütun

print("x",x,"\n")
print("y",y)
````

````python
A = np.array([[1, 2, 3],
              [1, 2, 3],
              [1, 2, 3]]) 

a = A + 1  # Eleman Eleman Toplama
b = A * 2  # Eleman Eleman Çarpma
c = A - 1  # Eleman Eleman Çıkarma
d = A / 2  # Eleman Eleman Bölme

print("a",a,"\n")
print("b",b,"\n")
print("c",c,"\n")
print("d",d,"\n")

e = A**2  # Eleman Eleman Üs Alma
f = A * A # Eleman Eleman Eşleştirerek Kendisiyle çarpma
print("e",e,"\n")
print("f",f,"\n")
````

* ``A**2`` ve ``A * A`` ifadeleri ile ``A @ A`` ifadesinin aşağıda verilen örnekte görüleceği üzere farklı sonuçları var.
* ``**`` ve ``*`` operatörler ile eleman eleman çarpma veya üst almas işlemi yapılırken ``@`` operatörü ile matris çarpımı yapılmıştır

````python
A = np.array([[1, 2, 3],
              [1, 2, 3],
              [1, 2, 3]])

a = A @ A
b = A * A
c = A**2
print(a)
print(b)
print(c)
````

### Matplotlib ile Grafik Çizim Mantığı
* Grafik, belirlediğimiz sonlu sayıdaki $(x, y)$ - $(x, y, z)$ koordinatlarından oluşur.
* Matplotlib, bu noktaları birbirine en kısa yoldan (düz çizgilerle) bağlar.
* Nokta sayısı ne kadar fazlaysa grafik okadar gerçek görünür.
* ``np.linspace(başlangıç,bitiş,başlangıç ve bitiş arasında kaç tane sayı olsun)`` komutu ile yeteri kadar sayıda eksen değeri oluşturusak grafik okadar pürüssüz görünür.
* Aşağıda sinüs grafiği üzerinden nokta sayısının grafiğe etkisini inceleyelim>

````python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.sin(x)

plt.figure(figsize=(16, 10))

# 1. Grafik
x1 = np.linspace(0, 10, 3)
plt.subplot(2, 3, 1)
plt.plot(x1, f(x1), 'r-o') 
plt.title("3 Nokta")

# 2. Grafik
x2 = np.linspace(0, 10, 5)
plt.subplot(2, 3, 2)
plt.plot(x2, f(x2), 'g-o')
plt.title("5 Nokta")

# 3. Grafik
x3 = np.linspace(0, 10, 10)
plt.subplot(2, 3, 3)
plt.plot(x3, f(x3), 'b-o')
plt.title("10 Nokta ")

# 4. Grafik
x4 = np.linspace(0, 10, 20)
plt.subplot(2, 3, 4)
plt.plot(x4, f(x4), 'm-o')
plt.title("20 Nokta")

# 5. Grafik
x5 = np.linspace(0, 10, 30)
plt.subplot(2, 3, 5)
plt.plot(x5, f(x5), 'k')
plt.title("30 Nokta ")

# 6. Grafik
x6 = np.linspace(0, 10, 1000)
plt.subplot(2, 3, 6)
plt.plot(x6, f(x6), 'y')
plt.title("1000 Nokta")

plt.suptitle("Linspace Nokta Sayısının Grafiğe Etkisi", fontsize=20)
plt.tight_layout()
plt.show()
````

# Gradient Descent ile iteratif olarak minimum noktası bulma

# Tek Değişkenli Fonksiyonlar İçin Gradient Descent
- $x_{k+1} = x_k - \alpha_k \frac{df}{dx}$
-  $x_{yeni} = x_{eski} - \alpha \frac{df}{dx}$


## $ f(x) = x^4 + 7*x^3 + 5*x^2 - 17*x + 3$
- X Başlagıç = -6 a = 0.001 e = 0.000001 maximum iterasyon = 1000 olsun

````python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return x**4 + 7*x**3 + 5*x**2 - 17*x + 3

def df(x):
    return 4*x**3 + 21*x**2 + 10*x - 17

eski_x = -6          # x'in başlangıç değeri
alpha = 0.001       # Adım uzunluğu (learning rate)
epsilon = 0.000001  # Durdurma kriteri
maxIterasyon = 1000 # Maksimum iterasyon sayısı


x_grafik = np.linspace(-6, 2, 1000)   
y_grafik = f(x_grafik)

plt.figure(figsize=(4, 4))
plt.plot(x_grafik, y_grafik, label='f(x)', color='blue', linewidth=2)
plt.title('Tek Değişkenli Fonksiyonlarda Gradient Descent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)


iterasyon = 0
while True:
    Df_eski_x = df(eski_x)
    yeni_x = eski_x - alpha * Df_eski_x  # Güncelleme denklemi
    
  
    Fxtem1 = f(eski_x)
    Fxtem2 = f(yeni_x)
    tempx = eski_x
    
    
    eski_x = yeni_x
    f_eski_x = f(eski_x)
    
    # DURDURMA KRİTERLERİ
    # 1. Fonksiyon değeri farkı < epsilon
    # 2. Adım büyüklüğü (L1 Normu) < epsilon -  güncel değişkenin eskinden çok farklı olmadığı belirtiliyor
    # 3. Maksimum iterasyon sayısı
    plt.plot(eski_x, f_eski_x, 'or', markersize=6, markerfacecolor='r', markeredgecolor='cyan')
    
    fark_vektoru = yeni_x - tempx
    if (abs(Fxtem1 - Fxtem2) <= epsilon) or (np.linalg.norm(fark_vektoru) <= epsilon) or (iterasyon == maxIterasyon):
        break
    
    iterasyon += 1

print(f"Algoritma {iterasyon}. iterasyonda durdu.")
print(f"Bulunan yerel minimum noktası x: {eski_x:.6f}")
print(f"Minimum f(x) değeri: {f_eski_x:.6f}")

plt.show()
````

# Gradyan Alanı Üzerinde Gradient Descent Adımları

````python
import matplotlib.pyplot as plt
import numpy as np

def fxy(x, y):
    return x**2 - 4*x + y**2 + 2*y

def grad_f(x, y):
    return np.array([2*x - 4, 2*y + 2])


x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

Z = fxy(X, Y)
U = 2*X - 4
V = 2*Y + 2

fig = plt.figure(figsize=(14, 7))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='inferno', alpha=0.6)
ax1.set_title("f(x,y) Yüzey Grafiği")
ax1.view_init(elev=45, azim=45)

ax2 = fig.add_subplot(122)
ax2.quiver(X, Y, U, V, color='gray', alpha=0.3, pivot='middle', label='Gradyan Alanı')
ax2.set_title("Gradyan Alanı Üzerinde İniş Adımları")


xeski = np.array([0.0, 1.0]) 
learning_rate = 0.1
epsilon = 0.01

for i in range(50):
    g = grad_f(xeski[0], xeski[1])
    xyeni = xeski - learning_rate * g
    
    # Quiver üzerinde adımı çiz: xeski'den xyeni'ye bir kırmızı ok
    ax2.annotate('', xytext=(xeski[0], xeski[1]), xy=(xyeni[0], xyeni[1]),
                 arrowprops=dict(arrowstyle="->", color='red', lw=1.5))
    
    # Noktayı işaretle
    ax2.scatter(xeski[0], xeski[1], color='blue', s=20, zorder=5)
    
    if np.linalg.norm(xyeni - xeski) < epsilon:
        break
    
    xeski = xyeni

# Minimum noktasını vurgula
ax2.scatter(xeski[0], xeski[1], color='green', s=100, marker='*', label='Bulunan Min', zorder=10)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
````
